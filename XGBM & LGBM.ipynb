# Building Predictive Models:

```python
x_train=train.iloc[:,1:12]
y_train=train.iloc[:,-1]
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.30, random_state=42)
```

```python
xgb =XGBClassifier(n_estimators=500,max_depth=5)

xgb.fit(x_train, y_train)
```

```python
y_pred = xgb.predict(x_test)
predictions = [round(value) for value in y_pred]
y_pred
```

```python
y_pred1 = xgb.predict(test)
predictions1 = [round(value) for value in y_pred1]
y_pred1
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
precision = precision_score(y_test, predictions)
recall = recall_score(y_test,predictions )
f1 = f1_score(y_test,predictions )

print(f'Precision: {precision}')
print(f' Recall: {recall}')
print(f'F1-score: {f1}')
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(x_train, y_train)
```

```python
params = {}
params['learning_rate'] = 0.002,
params['boosting_type'] = 'gbdt' 
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5 
params['num_leaves'] = 10
params['min_data'] = 20
params['max_depth'] = 10
clf = lgb.train(params, d_train, 150)
```

```python
y_pred3 = clf.predict(x_test)
predictions3 = [round(value) for value in y_pred3]
predictions3
```

```python
accuracy3 = accuracy_score(y_test, predictions3)
print("Accuracy: %.2f%%" % (accuracy3 * 100.0))
```

```python
precision = precision_score(y_test, predictions3)
recall = recall_score(y_test,predictions3 )
f1 = f1_score(y_test,predictions3 )

print(f'Precision: {precision}')
print(f' Recall: {recall}')
print(f'F1-score: {f1}')
```

```python
y_pred4 = clf.predict(test)
predictions4 = [round(value) for value in y_pred4]
predictions4
```

```python
xgb_param_grid = {'max_depth': [3, 5, 7],'learning_rate': [0.01, 0.05, 0.1],'n_estimators': [100, 200]}
xgb_grid = GridSearchCV(XGBClassifier(), xgb_param_grid, cv=5, scoring='accuracy')
xgb_grid.fit(x_train, y_train)
xgb_best_params = xgb_grid.best_params_
xgb_best_score = xgb_grid.best_score_

print("XGBoost Best Params: ", xgb_best_params)
print("XGBoost Best Score: ", xgb_best_score)
```

```python
lgb_param_grid = {'num_leaves': [31, 50],'learning_rate': [0.01, 0.05, 0.1],'n_estimators': [100, 200]}
lgb_grid = GridSearchCV(lgb.LGBMClassifier(), lgb_param_grid, cv=5, scoring='accuracy')
lgb_grid.fit(x_train, y_train)
lgb_best_params = lgb_grid.best_params_
lgb_best_score = lgb_grid.best_score_

print("LightGBM Best Params: ", lgb_best_params)
print("LightGBM Best Score: ", lgb_best_score)
```
